## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df['ConvertedComp'].plot(kind='box', figsize=(8, 6))
plt.title('Box plot of Converted Compensation')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
print("Give the five number summary for the column ConvertedComp:\n", df['ConvertedComp'].describe())

df['ConvertedComp'].dropna(axis=0, inplace=True)
Q1, Q3 = df['ConvertedComp'].quantile(.25), df['ConvertedComp'].quantile(.75)
IQR = Q3 - Q1
print('\nThe Inter Quartile Range for ConvertedComp:', IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
# your code goes here
upper = Q3 + (IQR * 1.5)
lower = Q1 - (IQR * 1.5)

print('Upper Bound:', upper)
print('Lower Bound:', lower)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here

outliers = (df['ConvertedComp'] < lower) | (df['ConvertedComp'] > upper)

print("number of outliers: " + str(len(outliers)))

# print("max outlier value: "+ str(outliers.max()))

# print("min outlier value: "+ str(outliers.min()))

outliers
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python

df2 = [(df['ConvertedComp'] < lower) | (df['ConvertedComp'] > upper)]
df[df2] = np.nan
print(df2)

print("number of outliers: " + str(len(df2)))
```
